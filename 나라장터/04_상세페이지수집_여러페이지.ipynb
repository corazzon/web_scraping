```python
import pandas as pd
```

```python
df = pd.read_csv("data/g2b_2020-09-06.csv")
df.shape
```

```python
df = df.head(10)
df.shape
```

```python
# 이전 실습에서 만든 bid_detail.py 파일을 import 해서 사용합니다.
# bid_detail.py 에 있는 함수를 불러와서 데이터를 수집할 예정입니다.
import bid_detail

# bid_detail.get_bid_info(bidno, bidseq)
```

```python
def get_bid_detail(bidno_seq):
    """공고번호-차수로 상세내역을 수집합니다."""
    bidno_seq_split = bidno_seq.split("-")
    bidno = bidno_seq_split[0]
    bidseq = bidno_seq_split[1]
    response = bid_detail.get_bid_info(bidno, bidseq)
    response = response.T
    response["공고번호-차수"] = bidno_seq
    return response
```

```python
# 일부 공고번호-차수를 가져와 제대로 수집이 되는지 확인합니다.
nums = df["공고번호-차수"].unique().tolist()
```

```python
get_bid_detail("20200905084-00")
```

```python
bidno_seq = "20200905275-00"
bid_detail_list = []
for bidno_seq in nums:
    print(bidno_seq)
    detail = get_bid_detail(bidno_seq)
    bid_detail_list.append(detail)
```

```python
df_detail = pd.concat(bid_detail_list)
df_detail.shape
```
