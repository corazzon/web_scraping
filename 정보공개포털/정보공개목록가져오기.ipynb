## 정보공개포털 정보목록 수집
* https://www.open.go.kr/othicInfo/infoList/infoList.do

```python
# url 인코딩을 해줍니다.
from urllib import parse

kwd = parse.quote("선불식 할부거래")
```

```python
# 복사해온 form data 를 분리합니다.
"kwd=%EA%B5%90%EC%9C%A1&insttSeCd=&eduYn=N&startDate=20200812&endDate=20200910&insttCdNm=&insttCd=&searchMainYn=&viewPage=1&rowPage=10&sort=s"
```

```python
#  form data 에 들어갈 값을 변경합니다.

base_url = "https://www.open.go.kr/othicInfo/infoList/infoList.ajax"
query = f"kwd={kwd}"
query = f"{query}&insttSeCd=&eduYn=N&startDate=20190112&endDate=20200910"
query = f"{query}&insttCdNm=&insttCd=&searchMainYn=&viewPage=1&rowPage=1000&sort=s"
```

```python
url = f"{base_url}?{query}"
url
```

## requests 로 데이터 요청하고 받아오기

```python
import requests

# post 로 요청을 보냅니다.
# response 라는 변수로 받습니다.
response = requests.post(url)
```

## 수집하고자 하는 데이터 찾기

```python
# 결과가 json 이기 때문에 json 으로 받습니다.
response_json = response.json()
```

```python
# 결과에서 원하는 값만 가져옵니다.
rtnList = response_json["result"]["rtnList"]
# 위에서 3개만 슬라이싱으로 가져옵니다.
rtnList[:3]
```

## 데이터프레임으로 만들기

```python
import pandas as pd
```

```python
# json의 key:value 데이터를 판다스의 데이터프레임 형식으로 변환해 줍니다.
df = pd.DataFrame(rtnList)
df.shape
```

```python
#  \t(탭문자)
#  \n(줄바꿈 문자) 를 공백문자로 변경합니다.
df["tma_kwd"] = df["tma_kwd"].str.replace("\n", " ")
```

## CSV형태로 저장하기

```python
# CSV 파일로 저장합니다.
df.to_csv(f"{kwd}.csv", encoding="cp949", index=False)
```
