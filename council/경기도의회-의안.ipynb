```python
from tqdm import tqdm, trange
tqdm.pandas()

page_list = []
for page_idx in trange(1, 10):
    url = find_url(page_idx)
    response = requests.get(url, verify=False)
    temp = pd.read_html(response.text)[0]
    sub_no = bs(response.text).select("td > a")
    sub_list_no = [sno["href"].split("/")[-1] for sno in sub_no]
    temp["내용번호"] = sub_list_no
    page_list.append(temp)
```

```python
df = pd.concat(page_list)
df.to_csv("경기도의안목록.csv", index=False)
pd.read_csv("경기도의안목록.csv")
```

```python
def find_sub_page(sub_no):
    sub_url = f"https://www.ggc.go.kr/site/agendaif/app/agndsrchList/DetailView/{sub_no}"
    response = requests.get(sub_url, verify=False)
    html = bs(response.text)
    table = html.select("div.view_cont > div > table")
    df_t = pd.read_html(str(table))
    df_sub_content = pd.concat([df_t[1].set_index(0).T,
                               df_t[-1].set_index(0).T], axis=1)
    df_sub_content["파일경로"] = str(html.select(".file_ico > a"))
    df_sub_content["내용번호"] = sub_no
    return df_sub_content
```

```python
df_sub = find_sub_page(sub_no=7798)
df_sub
```

```python
content_list = df["내용번호"].progress_map(find_sub_page)
```

```python
df_sub_list = pd.concat(row_list, axis=0, ignore_index=True)
```
