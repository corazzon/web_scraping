## 스타벅스 매장정보 
https://www.starbucks.co.kr/store/store_map.do?disp=quick

```python
import requests
```

```python
url = "https://www.starbucks.co.kr/store/getStore.do?r=RO98U559C0"
referer = "https://www.starbucks.co.kr/store/store_map.do?disp=quick"
params = "in_biz_cds=0&in_scodes=0&ins_lat=37.368665899999996&ins_lng=127.11512619999999&search_text=&p_sido_cd=01&p_gugun_cd=&isError=true&in_distance=0&in_biz_cd=&iend=1000&searchType=C&set_date=&rndCod=3WBZBCPSZ8&all_store=0&T03=0&T01=0&T12=0&T09=0&T30=0&T05=0&T22=0&T21=0&T10=0&T36=0&P10=0&P50=0&P20=0&P60=0&P30=0&P70=0&P40=0&P80=0&whcroad_yn=0&P90=0&new_bool=0"
```

```python
params_split = params.split("&")
params_list = {}
for p in params_split[1:]:
    p_split = p.split("=")
    key = p_split[0]
    val = p_split[-1]
    params_list[key] = val
    
params_list
```

```python
response = requests.post(url, params=params_list)
data = response.json()
data["list"][0]
```

```python
import pandas as pd
store = pd.DataFrame(data["list"])
print(store.shape)
store.head()
```

```python
store_null = store.isnull().sum()

null_cols = store_null[store_null == store.shape[0]].index.tolist()
print(null_cols)
```

```python
store_0 = (store == 0).sum()
store_0_cols = store_0[store_0 == store.shape[0]].index.tolist()
print(store_0_cols)
```

```python
null_cols.extend(store_0_cols)
print(null_cols)
```

```python
store = store.drop(null_cols, axis=1)
store.shape
```
