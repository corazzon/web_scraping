```python
url ="https://www.airportal.go.kr/life/airinfo/RaSkeDepList.jsp"
params = "gubun=c_getList&curr_page=1&one_page=10&one_group=10&dep_airport=RKSI&arr_airport=&srch_type=dep&current_dt_from=20210502&current_dt_to=20210502&regCls=0&al_icao=&fp_iata="
```

```python
# response = requests.post(url, params=params, headers=headers)
response = requests.post(url, params=params)
html = bs(response.text, "lxml")
```

```python
table = html.select("table")
table_info = table[0].select("table")[1]
table_info
```

```python
tds = table_info.select("td")
tds
```

```python
import re
re.findall("[0-9]{4}-[0-9]{2}-[0-9]{2}", "2020-01-01")
#     
```

```python

re.findall("[0-9]{4}-[0-9]{1,2}-[0-9]{1,2}", "2020-1-1")
```

```python
import re
td_list = []
tr = []
for td in tds:
    td_text = td.get_text()
    
    find_text = re.findall("[0-9]{4}-[0-9]{1,2}-[0-9]{1,2}", td_text)
    if len(find_text) > 0:
        tr = []
        td_list.append(tr)
        
    tr.append(td_text)
    print(td_text)
```
